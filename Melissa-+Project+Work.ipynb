```python
import pandas as pd
df = pd.read_csv("housingData.csv")
```

```python
import seaborn as sns
```

```python
sns.distplot(df['DIV'])
```

```python
%matplotlib inline
#graph sales price vs locating using regplot to make a scatterplot
sns.regplot(x= df['SLPR'], y = df['DIV'])
```

```python
#small multiples using bedrooms and bathrooms
import matplotlib.pyplot as plt

multiples= sns.FacetGrid(df, col= "DIV")
multiples = multiples.map(plt.hist, "SLPR")
```

```python
#create a model for sales price and location
import statsmodels.api as sm

Y = df["SLPR"]
X = df["DIV"]

model = sm.OLS(Y,X).fit()
print(model.summary())
```

```python
#small multiples using sales price and division
import matplotlib.pyplot as plt

multiples= sns.FacetGrid(df, col= "DIV", col_wrap=5)
multiples = multiples.map(plt.hist, "SLPR")
```

```python
#small multiples using sales price and division
import matplotlib.pyplot as plt

multiples= sns.FacetGrid(df, col= "DIV", col_wrap=3)
multiples = multiples.map(plt.hist, "SLPR")
plt.tight_layout()
```

```python
#look at metropolitan versus sales price
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(df["SLPR"], df["METRO"])
plt.xlabel("Sales Price")
plt.ylabel("Metro vs Non-Metro Area")
plt.title("Sales Price vs Metropolitan Area")
plt.tight_layout()
```

```python
#look at division versus sales price
import matplotlib.pyplot as plt
import numpy as np

plt.scatter(df["SLPR"], df["DIV"])
plt.xlabel("Sales Price")
plt.ylabel("Division")
plt.title("Sales Price vs Division")
plt.tight_layout() 
```
